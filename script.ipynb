```python
import matplotlib.pyplot as plt
import seaborn as sns

df_filtered = df_hotel[df_hotel['is_canceled'] == 0]

# Convertir los meses a un orden correcto (si no estÃ¡n ya ordenados)
month_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]
df_filtered['arrival_date_month'] = pd.Categorical(
    df_filtered['arrival_date_month'], categories=month_order, ordered=True
)

# Agrupar los datos por mes y tipo de hotel
occupancy_data = df_filtered.groupby(['arrival_date_month', 'hotel']).size().reset_index(name='reservations')

# Crear el grÃ¡fico
plt.figure(figsize=(12, 6))
sns.barplot(
    data=occupancy_data,
    x='arrival_date_month',
    y='reservations',
    hue='hotel',
    palette='viridis'
)

# Personalizar el grÃ¡fico
plt.title('OcupaciÃ³n Anual del Hotel por Mes y Tipo de Hotel', fontsize=16)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('NÃºmero de Reservas', fontsize=12)
plt.legend(title='Tipo de Hotel', loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar el grÃ¡fico
plt.show()

```
